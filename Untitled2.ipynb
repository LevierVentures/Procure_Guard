```python
## Build a summary KPI dashboard across all departments
def department_summary_dashboard(df):
    summary_list = []

    for department in sorted(df['Department'].unique()):
        dept_df = df[df['Department'] == department].copy()

        # Calculate Available Funds
        dept_df['Available Funds ($)'] = dept_df['PO Amount ($)'] - dept_df['Amount Spent ($)']
        dept_df['Available Funds (%)'] = (dept_df['Available Funds ($)'] / dept_df['PO Amount ($)']) * 100
        dept_df['Low Funds Flag'] = dept_df['Available Funds (%)'] < 10

        total_pos = len(dept_df)
        risky_count = dept_df['Low Funds Flag'].sum()
        risky_percentage = (risky_count / total_pos) * 100 if total_pos > 0 else 0
        average_available_funds = dept_df.loc[dept_df['Low Funds Flag'], 'Available Funds (%)'].mean() if risky_count > 0 else 0

        summary_list.append({
            'Department': department,
            'Total POs': total_pos,
            'Risky POs': risky_count,
            'Risk %': round(risky_percentage, 2),
            'Avg Funds Available (Risky)': round(average_available_funds, 2)
        })

    # Create a summary dataframe
    summary_df = pd.DataFrame(summary_list)

    # Sort by highest Risk % first
    summary_df = summary_df.sort_values('Risk %', ascending=False)

    display(summary_df)
```

```python
# Import again if needed
import pandas as pd

# Create the full KPI dashboard dataframe
def department_summary_dashboard_with_colors(df):
    summary_list = []

    for department in sorted(df['Department'].unique()):
        dept_df = df[df['Department'] == department].copy()

        # Calculate Available Funds
        dept_df['Available Funds ($)'] = dept_df['PO Amount ($)'] - dept_df['Amount Spent ($)']
        dept_df['Available Funds (%)'] = (dept_df['Available Funds ($)'] / dept_df['PO Amount ($)']) * 100
        dept_df['Low Funds Flag'] = dept_df['Available Funds (%)'] < 10

        total_pos = len(dept_df)
        risky_count = dept_df['Low Funds Flag'].sum()
        risky_percentage = (risky_count / total_pos) * 100 if total_pos > 0 else 0
        average_available_funds = dept_df.loc[dept_df['Low Funds Flag'], 'Available Funds (%)'].mean() if risky_count > 0 else 0

        summary_list.append({
            'Department': department,
            'Total POs': total_pos,
            'Risky POs': risky_count,
            'Risk %': round(risky_percentage, 2),
            'Avg Funds Available (Risky)': round(average_available_funds, 2)
        })

    # Create the DataFrame
    summary_df = pd.DataFrame(summary_list)
    summary_df = summary_df.sort_values('Risk %', ascending=False)

    # Apply color styling
    styled_df = summary_df.style.background_gradient(subset=['Risk %'], cmap='Reds').format({
        'Risk %': "{:.2f}",
        'Avg Funds Available (Risky)': "{:.2f}"
    })

    return styled_df
```

```python
import pandas as pd
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# --- Settings ---
sender_email = "enlightenize0@gmail.com"
app_password = "ycbdnzteceusydhs"  # your app password (no spaces)
receiver_email = "enlightenize0@gmail.com"  # later replace per department owner

# --- Load your dataset ---
df = pd.read_csv('po_dataset_1000.csv')

# --- Calculate Available Funds and Risk Flag ---
df['Available Funds ($)'] = df['PO Amount ($)'] - df['Amount Spent ($)']
df['Available Funds (%)'] = (df['Available Funds ($)'] / df['PO Amount ($)']) * 100
df['Low Funds Flag'] = df['Available Funds (%)'] < 10

# --- Email function ---
def send_risky_pos_email(department_name, risky_pos_df):
    subject = f"⚠️ PO Low Funds Alert - {department_name} Department"

    body = f"""
    <p>Hello,</p>
    <p>The {department_name} department has the following POs with <b>less than 10% Available Funds</b>:</p>
    <table border="1" cellpadding="5" cellspacing="0">
        <tr>
            <th>PO Number</th>
            <th>Supplier</th>
            <th>PO Amount ($)</th>
            <th>Amount Spent ($)</th>
            <th>Available Funds ($)</th>
            <th>Available Funds (%)</th>
        </tr>
    """
    for _, row in risky_pos_df.iterrows():
        body += f"""
        <tr>
            <td>{row['PO Number']}</td>
            <td>{row['Supplier']}</td>
            <td>${row['PO Amount ($)']:.2f}</td>
            <td>${row['Amount Spent ($)']:.2f}</td>
            <td>${row['Available Funds ($)']:.2f}</td>
            <td>{row['Available Funds (%)']:.2f}%</td>
        </tr>
        """

    body += """
    </table>
    <p>Please review and address these POs urgently.</p>
    <p>Regards,<br>PO Monitoring System</p>
    """

    # Create email
    msg = MIMEMultipart("alternative")
    msg['From'] = sender_email
    msg['To'] = receiver_email
    msg['Subject'] = subject

    msg.attach(MIMEText(body, 'html'))

    # Send email
    try:
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.starttls()
        server.login(sender_email, app_password)
        server.send_message(msg)
        server.quit()
        print(f"✅ Email sent for {department_name} Department ({len(risky_pos_df)} risky POs).")
    except Exception as e:
        print(f"❌ Failed to send email for {department_name}: {e}")

# --- Check and Send for Each Department ---
departments = df['Department'].unique()

for department in departments:
    dept_df = df[df['Department'] == department]
    risky_pos_df = dept_df[dept_df['Low Funds Flag']]

    if not risky_pos_df.empty:
        send_risky_pos_email(department, risky_pos_df)

print("✅ Monitoring Completed for all departments.")
```
